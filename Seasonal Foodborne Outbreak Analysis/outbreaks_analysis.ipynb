```python
# Create a mapping dictionary for months
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

# Convert the "Month" column to integers
outbreaks_data['Month'] = outbreaks_data['Month'].map(month_mapping)
```

```python
# Filter the data to include only relevant columns and non-null values
data = outbreaks_data['Month'].dropna()

# Compute the CDF
cdf = thinkstats2.Cdf(data)

# Plot the CDF
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Month', ylabel='CDF', title='CDF of Foodborne Illnesses by Month')

# Show the plot
thinkplot.show()
```

```python
# The CDF shows how the cumulative probability of foodborne illnesses changes as we move through the months.
# If the curve is steeper in certain months, it suggests a higher prevalence of foodborne illnesses during those 
# months compared to others. If the curve is less steep, it indicates a lower prevalence of foodborne illnesses 
# during those months. 

# As shown above, the CDF shows a slightly steeper (blink and you'll miss it) curve during certain months
# (e.g., spring/summer months), this would suggests there is a higher prevalence of foodborne illnesses during 
# those seasons. Overall, there are minimal changes through the months, indicating a more consistent 
# prevalence of foodborne illnesses throughout the year.
```

# Create two scatter plots comparing two variables and provide your analysis on correlation and causation. Remember, covariance, Pearsonâ€™s correlation, and Non-Linear Relationships should also be considered during your analysis (Chapter 7).

```python
# Remove rows with missing values
outbreaks_data = outbreaks_data[['Illnesses', 'Hospitalizations']].dropna() 
```

```python
# Select the variables of interest
illnesses = outbreaks_data['Illnesses']
hospitalizations = outbreaks_data['Hospitalizations']

# Create Scatter plot
plt.scatter(illnesses, hospitalizations)
plt.xlabel('Reported Illnesses')
plt.ylabel('Hospitalizations')
plt.title('Scatter Plot of Illnesses vs Hospitalizations')
plt.show()

# Calculate Correlation analysis
correlation = np.corrcoef(illnesses, hospitalizations)[0,1]
print('Correlation coefficient:', correlation)

# Causation analysis
covariance = np.cov(illnesses, hospitalizations)[0,1]
print('Covariance:', covariance)

# Pearson's correlation
pearson_corr = illnesses.corr(hospitalizations)
print("Pearson's correlation coefficient:", pearson_corr)

# Non-linear relationship analysis
sns.regplot(x=illnesses, y=hospitalizations, scatter_kws={'alpha':0.5, 'color':'blue'})
plt.xlabel('Reported Illnesses')
plt.ylabel('Hospitalizations')
plt.title('Scatter Plot with Regression Line')
plt.show()
```

# Conduct a test on your hypothesis using one of the methods covered in Chapter 9.

```python
# Import libraries
import pandas as pd
import numpy as np

# Read csv
outbreaks_data = pd.read_csv('FoodborneOutbreaks.csv')
```

```python
# Extract the relevant columns from the dataset
months = outbreaks_data['Month']
cases = outbreaks_data['Illnesses']

# Create a boolean mask to identify summer months (June, July, August)
is_summer = (months.isin(['June', 'July', 'August']))

# Separate the cases for summer and non-summer months
summer_cases = cases[is_summer]
non_summer_cases = cases[~is_summer]
```

```python
# Calculate observed test statistic
observed_statistic = np.mean(summer_cases) - np.mean(non_summer_cases)
```

```python
# Perform permutation test
num_permutations = 1000
test_statistics = []

for _ in range(num_permutations):
    # Shuffle the summer and non-summer cases
    shuffled_cases = np.random.permutation(cases)

    # Calculate the test statistic using shuffled cases
    shuffled_summer_cases = shuffled_cases[is_summer]
    shuffled_non_summer_cases = shuffled_cases[~is_summer]
    test_statistic = np.mean(shuffled_summer_cases) - np.mean(shuffled_non_summer_cases)

    test_statistics.append(test_statistic)
```

```python
# Calculate the p-value
p_value = (np.abs(test_statistics) >= np.abs(observed_statistic)).mean()
```

```python
# Interpret results
alpha = 0.05  # significance level

if p_value < alpha:
    print("There is evidence to reject the null hypothesis.")
else:
    print("There is not enough evidence to reject the null hypothesis.")
```
