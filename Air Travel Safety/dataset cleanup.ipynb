# Initial Dataset 2: Accidents & Fatalities Per Year

```python
# View df
accidents_df.head()
```

```python
# Correcting the renaming and cleaning steps
accidents_df.columns = [
    'Year',
    'Accidents_Excluding_Extreme_Events',
    'Fatalities_Excluding_Extreme_Events',
    'Unnamed_3', # Placeholder, to be removed
    'Accidents_Including_Extreme_Events',
    'Fatalities_Including_Extreme_Events',
    'Unnamed_6', # Placeholder, to be removed
    'Accidents_Passenger_Flights_Including_Extreme_Events',
    'Fatalities_Passenger_Flights_Including_Extreme_Events',
    'Unnamed_9', # Placeholder, to be removed
    # The rest will be adjusted accordingly
    'Unnamed_10', 'Unnamed_11', 'Unnamed_12', 'World_Air_Traffic_Departures', 
    'Unnamed_14', 'Unnamed_15', 'Unnamed_16', 'Five_Year_Moving_Avg', 
    'Unnamed_18', 'Unnamed_19', 'Civilian_Corporate_Jets_Accidents', 
    'Civilian_Corporate_Jets_Fatalities', 'Moving_5_Year_Average_Accidents']
```

```python
# Dropping the unnecessary 'Unnamed' placeholder columns
columns_to_drop = ['Unnamed_3', 'Unnamed_6', 'Unnamed_9', 'Unnamed_10', 'Unnamed_11', 'Unnamed_12', 'Unnamed_14', 'Unnamed_15', 'Unnamed_16', 'Unnamed_18', 'Unnamed_19']
accidents_cleaned = accidents_df.drop(columns=columns_to_drop)
```

```python
# View rows to confirm the structure
accidents_cleaned = accidents_cleaned.iloc[1:]

# Reset index & view cleaned data
accidents_cleaned.reset_index(drop=True, inplace=True)
accidents_cleaned.head()
```

```python
# Filling numerical blanks with 0
for column in accidents_cleaned.columns[1:]:
        accidents_cleaned[column].fillna(0, inplace=True)
else:
        accidents_cleaned[column].fillna('Unknown', inplace=True)

# Check for any remaining null values
remaining_nulls = accidents_cleaned.isnull().sum().sum()
remaining_nulls
```

```python
# Save the cleaned dataset to a new CSV file
accidents_cleaned.to_csv('cleaned_accidents.csv', index=False)
```

# Supplemental Dataset 3: Airplane Crashes & Fatalities Since 1908

```python
# View df
airplane_crash_df.head()
```

```python
# Get summary of df
print(airplane_crash_df.info())
```

```python
# Get stats of df
print(airplane_crash_df.describe())
```

```python
# Check for duplicate rows
print(airplane_crash_df.duplicated().sum())
```

```python
# Fill missing 'Time' with 'Unknown'
airplane_crash_df['Time'] = airplane_crash_df['Time'].fillna('Unknown')

# Fill missing numerical values ('Aboard', 'Fatalities', 'Ground') with 0
airplane_crash_df[['Aboard', 'Fatalities', 'Ground']] = airplane_crash_df[['Aboard', 'Fatalities', 'Ground']].fillna(0)

# Convert 'Date' to datetime format
airplane_crash_df['Date'] = pd.to_datetime(airplane_crash_df['Date'], errors='coerce')

# Creating new 'Year' column from the 'Date' column for trend analysis
airplane_crash_df['Year'] = airplane_crash_df['Date'].dt.year

# Impute missing values with 'Unknown'
airplane_crash_df['Location'].fillna('Unknown', inplace=True)
airplane_crash_df['Operator'].fillna('Unknown', inplace=True)
airplane_crash_df['Route'].fillna('Unknown', inplace=True)
airplane_crash_df['Type'].fillna('Unknown', inplace=True)
airplane_crash_df['Registration'].fillna('Unknown', inplace=True)
airplane_crash_df['cn/In'].fillna('Unknown', inplace=True)
airplane_crash_df['Summary'].fillna('Unknown', inplace=True)

# Drop 'Flight #' column
airplane_crash_df.drop(columns=['Flight #'], inplace=True)
```

```python
# Check for missing values
print(airplane_crash_df.isnull().sum())
```

```python
# Save the cleaned dataset to a new CSV file
airplane_crash_df.to_csv('cleaned_airplane_crash.csv', index=False)
```
